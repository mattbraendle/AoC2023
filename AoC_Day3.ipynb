1. Assign a value to `nums`

**AI Prompt**

Yes expand on the walrus operator and why it is needed in addition to =, what is the difference?

**AI Response**

The key difference between `=` and `:=` is where they can be used:

1. Regular assignment `=`:
   - Can only be used as a standalone statement
   - Cannot be used inside expressions
```python
# This works
x = 5
if x > 3:
    print(x)

# This doesn't work
if (x = 5) > 3:  # Syntax Error!
    print(x)
```

2. Walrus operator `:=`:
   - Can be used inside expressions
   - Returns the assigned value
```python
# This works
if (x := 5) > 3:
    print(x)  # x is now 5

# Common use in while loops
while (line := input()) != "quit":
    print(f"You entered: {line}")
```

The walrus operator is particularly useful when:
1. You need to use a value multiple times
2. You want to assign and test in one step
3. You're working with list comprehensions or generator expressions

h*(T-h) > D

This expands to:
hT - h² > D

For T=7 and D=9, we need to solve:

Let's rearrange this into standard quadratic form:
-h² + 7h - 9 > 0

To find the boundaries of where this inequality is true, we first solve the equation:
-h² + 7h - 9 = 0

Would you like to help solve this quadratic equation? We can use the quadratic formula: (-b ± √(b² - 4ac))/(2a)
Where a=-1, b=7, and c=-9

Can you plug these values into the formula?

````python
import math
def quadratic_equation(a,b,c):
    d = b*b-4*a*c
    if d >= 0:
        x1 = (-b + math.sqrt(d))/(2*a)
        x2 = (-b - math.sqrt(d))/(2*a)
        return (x1,x2)
    else: return -1

res = quadratic_equation(-1,7,-9)
print(res)
res==-1
````

How to round numbers in python, up and down?

**AI Response**

In Python, there are several ways to round numbers:

1. `math.ceil(x)` - rounds up to the nearest integer
2. `math.floor(x)` - rounds down to the nearest integer
3. `round(x)` - rounds to the nearest integer (with ties going to the nearest even number)

For example:
```python
import math
x = 1.7
print(f"Original: {x}")
print(f"Ceiling: {math.ceil(x)}")
print(f"Floor: {math.floor(x)}")
```

Would you like to try using these functions with your quadratic equation solutions?

````python
def get_ints_in_interval(x1,x2,upper_boundary):
    x1=math.ceil(x1)
    x2=min(math.floor(x2),upper_boundary)
    if x2>=x1:
        return(list(range(x1,x2+1)))
    else: return -1

print(get_ints_in_interval(1.7,5.3,7))
````

````python
def get_max_per_color(game):
    max_colors = {'red': 0, 'green': 0, 'blue': 0}
    for entry in game:
        entry_list = entry.split(',')
        for s in entry_list:
            number, color = s.split()  # This gives you ['13', 'red'] as number, color
            max_colors[color]=max(int(number),max_colors[color]) # Now you need to update max_colors[color] if int(number) is larger...
    return max_colors

get_max_per_color(sample_list)
````
